# Preparation

Berikut adalah tahapan yang akan dilalui pada data `SC_HW1_bank_data.csv` sebelum tahap pemodelan :

## Import Library

```python
import pandas as pd
import numpy as np
```

## Input Dataset

```python
#Membaca data dan memasukkannya ke dalam bentuk Pandas Dataframe
df = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/SC_HW1_bank_data.csv')
```

```python
#Jalankan code untuk mengecek nama kolom yang tersedia
df.columns
```

```python
df.head()
```

```python
#Hilangkan kolom yang dirasa tidak relevan dengan model (contoh: ID). None dapat diisi dengan nama-nama kolom yang akan digunakan.
#Contoh df = df[['X1','X2', 'Y']].copy()

#START CODE
df = df[['Geography', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary', 'Exited']].copy()
```

## Preprocessing

```python
df['Geography'].unique()
```

```python
#Lakukan One-Hot Encoder untuk data categorical, dengan fungsi pandas get_dummies

#START CODE
df = pd.get_dummies(df)
df.head()
```

```python
#Pisahkan mana X (feature) dengan Y,
#Y adalah kolom "Exited"

#START CODE
X = df.drop('Exited', axis=1)
y = df['Exited']
```

```python
#Lakukan Scaler dan/atau Noermalisasi jika diperlukan
from sklearn.preprocessing import MinMaxScaler

#START CODE
scaler = MinMaxScaler()
X_transform = scaler.fit_transform(X)
X_transform
```

```python
#Ini digunakan jika dilakukan scaler/Normalisas. Jika tidak, code ini bisa dilewat dan diganti dengan code yang ada di dalam komen
X_transform = pd.DataFrame(X_transform, columns = X.columns)
X_transform
#X_transform = X.copy()
```

## Train-Test Split

```python
#Split menjadi train dan test dengan test_size 25%
#Tidak perlu mengubah code ini

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_transform,y,test_size = 0.25,random_state = 123)
```

# Modeling

## Model1

### Soal :
Jelaskan secara Singkat Model pertama yang digunakan!

Model yang pertama digunakan adalah Logistic Regression. Model ini digunakan untuk tugas klasifikasi, khususnya ketika Anda ingin memprediksi probabilitas terjadinya suatu peristiwa. Logistic Regression adalah salah satu algoritma klasifikasi yang paling sederhana dan umum digunakan. Model ini memodelkan hubungan antara variabel input dan probabilitas output dengan menggunakan fungsi logistik. Parameter yang diuji berupa 'C' (parameter regulasi), 'penalty' (jenis regulasi), 'solver' (algoritma penyelesaian), dan 'max_iter' (jumlah iterasi maksimum).

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}


#START CODE
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
params1 = {
    'C': [0.01, 0.1, 1, 10, 100],  # Invers dari regulasi (semakin kecil semakin kuat regulasi)
    'penalty': ['l1', 'l2'],  # Jenis regulasi (L1 atau L2)
    'solver': ['liblinear', 'saga'],  # Solver yang digunakan
    'max_iter': [100, 200, 300],  # Jumlah iterasi maksimum
}
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(
             estimator= model1,
             param_grid= params1,
             scoring = 'accuracy',
             n_jobs = 10, # core cpu yang digunakan
             cv = 10 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid.fit(X_train,y_train)
grid.best_params_

#1 menit 29 detik
```

```python
#lakukan evaluasi dengan beberapa maetric di bawah ini
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid.predict(X_test)

print("Classification Report : ")
print(classification_report(y_test,y_pred))

print("\nConfusion Matrix : ")
print(confusion_matrix(y_test,y_pred))

print("\nAccuracy Score : ")
print(accuracy_score(y_test, y_pred))
```

## Model2
### Soal :
Jelaskan secara Singkat Model ke-2 yang digunakan!

Model kedua yang digunakan adalah K-Nearest Neighbors (KNN): KNN adalah algoritma klasifikasi yang sederhana dan intuitif. Ide dasarnya adalah bahwa suatu data point diklasifikasikan berdasarkan mayoritas kelas dari tetangga terdekatnya dalam ruang fitur. Jadi, jika suatu data point memiliki tetangga-tetangga yang mayoritasnya termasuk dalam kelas A, maka data point tersebut akan diklasifikasikan ke dalam kelas A. Parameter yang diuji berupa 'n_neighbors' (jumlah tetangga terdekat yang akan dipertimbangkan), 'weights' (metode weighting), dan 'p' (parameter jarak).

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}


#START CODE
from sklearn.neighbors import KNeighborsClassifier
model2 = KNeighborsClassifier()
params2 = {
    'n_neighbors': [3, 5, 7],  # Jumlah tetangga terdekat yang akan dipertimbangkan
    'weights': ['uniform', 'distance'],  # Metode weighting (seragam atau berdasarkan jarak)
    'p': [1, 2]  # Parameter jarak (1 untuk jarak Manhattan, 2 untuk jarak Euclidean)
}
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV
grid2 = GridSearchCV(
             estimator= model2,
             param_grid= params2,
             scoring = 'accuracy',
             n_jobs = 10, # core cpu yang digunakan
             cv = 10 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid2.fit(X_train,y_train)
grid2.best_params_

#19 detik
```

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid2.predict(X_test)

print("Classification Report : ")
print(classification_report(y_test,y_pred))

print("\nConfusion Matrix : ")
print(confusion_matrix(y_test,y_pred))

print("\nAccuracy Score : ")
print(accuracy_score(y_test, y_pred))
```

## Model3
### Soal :
Jelaskan secara Singkat Model ke-3 yang digunakan!

Model ketiga yang digunakan adalah Support Vector Machine (SVM): SVM adalah algoritma klasifikasi yang kuat dan serbaguna yang digunakan untuk masalah klasifikasi dan regresi. SVM berusaha untuk menemukan hyperplane terbaik (garis atau permukaan) yang memisahkan dua kelas dalam ruang fitur. SVM dapat digunakan untuk masalah klasifikasi dengan dua kelas atau lebih, dan juga dapat diterapkan dalam ruang berdimensi tinggi. Parameter yang diuji berupa 'C' (parameter regulasi), 'kernel' (jenis kernel), dan 'gamma' (kernel coefficient).

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}


#START CODE
from sklearn.svm import SVC
model3 = SVC()
params3 = {
    'C': [0.1, 1, 10],  # Parameter regulasi
    'kernel': ['linear', 'rbf', 'poly'],  # Jenis kernel (linear, radial basis function, polynomial)
    'gamma': ['scale', 'auto', 0.1, 1],  # Kernel coefficient
}
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV
grid3 = GridSearchCV(
             estimator= model3,
             param_grid= params3,
             scoring = 'accuracy',
             n_jobs = 10, # core cpu yang digunakan
             cv = 10 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid3.fit(X_train,y_train)
grid3.best_params_
```

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid3.predict(X_test)

print("Classification Report : ")
print(classification_report(y_test,y_pred))

print("\nConfusion Matrix : ")
print(confusion_matrix(y_test,y_pred))

print("\nAccuracy Score : ")
print(accuracy_score(y_test, y_pred))
```
